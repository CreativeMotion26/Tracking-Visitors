```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime
```

```python
csv_file = 'data_log.csv'
```

```python
from IPython.core.logger import time
#Read data from CSV and store it in lists
timestamps = []
people_counts = []
```

```python
with open(csv_file, 'r')as file:
  reader = csv.reader(file)
  for row in reader:
    if len(row) == 2:
      timestamp_str, people_count_str = row
      try:
          timestamp = datetime.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S')
          people_count = float(people_count_str)
          timestamps.append(timestamp)
          people_counts.append(people_count)
      except ValueError:
        pass
```

```python
plt.figure(figsize=(12, 6))
plt.plot(timestamps, people_counts, marker='o', linestyle='-',color='b')
plt.xlabel('Time')
plt.ylabel('People Count')
plt.title('People count over time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
